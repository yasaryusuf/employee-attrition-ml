```python
import numpy as np
import pandas as pd

np.random.seed(42)
n = 500

age = np.random.randint(22, 60, n)
years_at_company = np.clip((age - 22) * np.random.uniform(0.2, 0.6, n), 0, 40).astype(int)
salary = np.clip(7000 + years_at_company * 900 + np.random.normal(0, 1500, n), 7000, 30000).astype(int)
overtime = np.random.binomial(1, 0.35, n)
job_satisfaction = np.random.randint(1, 6, n)

resign_prob = (
    0.35 * (job_satisfaction <= 2).astype(int) +
    0.25 * overtime +
    0.25 * (years_at_company < 2).astype(int)
)
resigned = np.random.binomial(1, np.clip(resign_prob, 0, 0.9))

df = pd.DataFrame({
    "age": age,
    "salary": salary,
    "years_at_company": years_at_company,
    "overtime": overtime,
    "job_satisfaction": job_satisfaction,
    "resigned": resigned
})

df.to_csv("data.csv", index=False)
print("data.csv oluşturuldu (500 satır)")
```
